## TSNE visulaization for UNSW

```python
def aggregate_data(data, selected):
    aggregation_dict = {}
    for feature in selected + ['label']:
        if 'label' in feature:
            aggregation_dict[feature] = 'max'
        elif 'num' in feature:
            aggregation_dict[feature] = lambda x: pd.Series.mode(x)[0]
        else:
            aggregation_dict[feature] = 'mean'
    agg_data = data.groupby('src_ip').agg(aggregation_dict).reset_index()
    return agg_data
```

```python
src_ips = ['59.166.0.' + str(i) for i in range(0,10)] + ['175.45.176.' + str(i) for i in range(0,4)]
features = ['src_port', 'dst_port', 'protocol_num', 'src_bytes', 'dst_bytes']
all_data = pd.DataFrame()
for i in range(1, 5):
    normal = pd.read_pickle('Datasets/UNSW-NB15/UNSW-NB15-' + str(i) + '/normal.pkl')
    normal = normal[normal['src_ip'].isin(src_ips)]
    anomalous = pd.read_pickle('Datasets/UNSW-NB15/UNSW-NB15-' + str(i) + '/anomalous.pkl')
    anomalous = anomalous[anomalous['src_ip'].isin(src_ips)]
    data = pd.concat([normal, anomalous], ignore_index=True).sort_values(by='date').reset_index(drop=True)
    data['src_ip'] = data['src_ip'] + '_' + str(i)
    all_data = pd.concat([all_data, data], ignore_index=True).sort_values(by='date').reset_index(drop=True)
agg_data = aggregate_data(all_data, features)
```

### Plot the heatmap of the aggregated view

```python
plt.figure(figsize=(10,6))
corr = agg_data.corr()
cmap = sns.color_palette("RdBu", 20)
sns.heatmap(corr, cmap=cmap, square=True, xticklabels=corr.columns.values, yticklabels=corr.columns.values, linewidths=0.5, cbar_kws={"shrink": .5})
```

```python
plt.figure(figsize=(10,6))
cmap = sns.color_palette("RdBu", 20)
scaled_df = pd.DataFrame(MinMaxScaler().fit_transform(agg_data[features + ['label']].values), index=agg_data.index, columns=features + ['label'])
heat = sns.heatmap(scaled_df, cmap=cmap, linewidths=0.3, cbar_kws={"shrink": .5}, robust=True)
heat.set_xticklabels(heat.get_xticklabels(), rotation=45)
```

### Plot the data in a lower space through TSNE

```python
agg_data.loc[agg_data['label'] == 0, 'label'] = 'Benign'
agg_data.loc[agg_data['label'] == 1, 'label'] = 'Malicious'
tsne_results = TSNE(n_components=2).fit_transform(agg_data[features].values)
agg_data['TSNE-2D 1st Dimension'] = tsne_results[:,0]
agg_data['TSNE-2D 2nd Dimension'] = tsne_results[:,1]
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="TSNE-2D 1st Dimension", y="TSNE-2D 2nd Dimension",
    hue="label",
    data=agg_data,
    palette=['red','blue'],
    legend="full",
    alpha=0.7
)
```
